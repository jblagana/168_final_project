```python
def show_caps():
    cap_locations = {}
    cap_name = dss.Capacitors.First()

    while cap_name:
        cap_loc = dss.CktElement.BusNames()[0]
        cap_locations[cap_name] = cap_loc

        cap_name = dss.Capacitors.Next()

    for cap, bus in cap_locations.items():
        print(f"Capacitor: {cap}, Bus Location: {bus}")
```

```python
node_voltages = dss.Circuit.AllBusMagPu()
print(node_voltages)
```

```python
def edit_cap_loc(new_buses):
    # Set the first capacitor active and get its name
    cap_names = dss.Capacitors.AllNames()

    for i, cap in enumerate(cap_names):
        # Set the new bus of the active capacitor
        dss.Text.Command(f'Edit Capacitor.{cap} bus1={new_buses[i]}')

```

```python
str_gene_space = dss.Circuit.AllNodeNames()
gene_space = [list(range(len(str_gene_space)))]*2
```

```python
def fitness_func():
    nominal_voltage = 1
    node_voltages = dss.Circuit.AllBusMagPu()
    # Calculate the mean squared error
    mse = sum((v - nominal_voltage) ** 2 for v in node_voltages) / len(node_voltages)
    return mse
```

```python
show_caps()
print(dss.Circuit.AllBusMagPu())
print(fitness_func())
print('-------------------------')

new_buses = ['684.3', '611.2']
edit_cap_loc(new_buses)
dss.Solution.Solve()
show_caps()
print(dss.Circuit.AllBusMagPu())
print('Fitness', fitness_func())
```

```python
import itertools

# Your list
lst = dss.Circuit.AllNodeNames()

# Generate permutations
perms = list(itertools.permutations(lst, 2))
max_fitness = -float('inf')
best_perm = ''

for perm in perms:
    edit_cap_loc(perm)
    dss.Solution.Solve()
    show_caps()
    print(dss.Circuit.AllBusMagPu())
    fitness = fitness_func()
    print('Fitness', fitness)
    if fitness > max_fitness:
        max_fitness = fitness
        best_perm = perm

print("Best permutation:", best_perm, "Fitness", max_fitness)
```
